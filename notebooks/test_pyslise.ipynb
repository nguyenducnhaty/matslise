```python
import pyslise as ps
```

```python
M = 6
p = ps.Pyslise(lambda x: x*x, 0, M, 32)
```

```python
E = 3
m = 1
f = p.propagate(E, [0,1], 0, m)[0][0]/p.propagate(E, [0,1], M, m)[0][0]
plot(lambda x: p.propagate(E, [0,1], 0, x)[0][0], (0,m)) + \
plot(lambda x: f*p.propagate(E, [0,1], M, x)[0][0], (m,M), color="red")
```

```python
p.computeEigenvaluesByIndex(0,10, [0,1],[0,10])
```
