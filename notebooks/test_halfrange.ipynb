```python
f = fast_callable((1-cos(2*pi*x))/2*1000, vars=["x"], domain=float)
```

```python
plot(f, (-1, 1))
```

```python
from pyslise import Pyslise
import numpy as np
```

```python
ps = Pyslise(f, 0, 1, 31)
sides = [[0,-1], [1,0]]
eigenvalues = ps.computeEigenvaluesByIndex(0,20, *sides)
eigenvalues
```

```python
xs = np.linspace(0,1,1000)
v, dv = ps.computeEigenfunction(eigenvalues[12][1], *sides, xs=xs)
list_plot(zip(xs, v))
```
